### Assignment 1: Pivot Table and Groupby (Etsy)
Create a pivot table to show the average product price by seller name and tags. Then, use the groupby function to group the data by seller name and calculate the total views and favorites for each seller.


```python
import pandas as pd
```

```python
itsy = pd.read_csv("etsy.csv")
itsy.head()
```

```python
pd.pivot_table(itsy, values="product_price", index=["seller_name","tags"])
```

```python
itsy.groupby(["seller_name"])[["seller_reviews","favorites"]].sum()
```
