### 3. Apply Function
In the Olympics dataset, write a function to categorize athletes into age groups (<20, 20-30, >30) and use the .apply() function to create a new column called Age_Group.

```python
def age_group(num):
    if num < 20:
        return "Young"
    elif num >= 30:
        return "Mature"
    elif num >= 20 and num < 30:
        return "Mid Age"
```

```python
olympics["Age_Group"] = olympics["Age"].apply(age_group)
```
