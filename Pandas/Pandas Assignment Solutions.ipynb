```python
import pandas as pd
```

```python
df = pd.read_csv("olympics.csv")
df.head()
```

```python
df.groupby(["Team"])["Age"].mean().sort_values(ascending=False)
```

```python
df.drop_duplicates("Name")
```

```python
def age(num):
    if num < 25:
        return "Young"
    elif 25 <= num <= 35:
        return "Mid"
    else:
        return "Senior"

df["Age"].apply(age)
```
